```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

df1 = pd.read_csv('/Users/rayhaan/Desktop/Personal Projects /Clinical Note Classification/data/Medical-Abstracts-TC-Corpus-main/medical_tc_train.csv')
df2 = pd.read_csv('/Users/rayhaan/Desktop/Personal Projects /Clinical Note Classification/data/Medical-Abstracts-TC-Corpus-main/medical_tc_test.csv')

print(df1.shape) #how many sample are contained in the data 
print(df1.columns) #What are the columns

X_train = df1["medical_abstract"]
Y_train = df1["condition_label"]
X_test = df2["medical_abstract"]
Y_test = df2["condition_label"]

tfidf = TfidfVectorizer(
    stop_words="english", 
    max_features=100000,)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print("X_train shape =", X_train_tfidf.shape)
print("X_test shape =", X_test_tfidf.shape)
print("Vocab Size =", len(tfidf.vocabulary_))

#Model Training 
model = LogisticRegression(max_iter= 5000, random_state=42)
model.fit(X_train_tfidf, Y_train)

#Model testing & accuracy scoring 
y_predict = model.predict(X_test_tfidf)
accuracy = accuracy_score(Y_test, y_predict)
print('Accuracy =', accuracy)
print(classification_report(Y_test, y_predict))
```
